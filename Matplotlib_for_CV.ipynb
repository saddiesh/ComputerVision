```python
im = array(Image.open("/Users/stephaniexia/Documents/didi/yinyuehui.jpg").convert('L'))
figure()
#contour: change to 'L' grey scale first
subplot(131)
gray()
contour(im, origin='image')
title("contour")

#histogram:
im = array(Image.open("/Users/stephaniexia/Documents/didi/yinyuehui.jpg"))
subplot(132)
print(im.flatten())
hist(im.flatten(), 256)
title("histogram")
show()

#histogram:
im = array(Image.open("/Users/stephaniexia/Documents/didi/yinyuehui.jpg").convert('L'))
subplot(133)
hist(im.flatten(), 256)
title("histogram_convert'L'")
show()
```
