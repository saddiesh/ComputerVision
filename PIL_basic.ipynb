```python
#To load an image from a file, use the open function in the Image module.

from PIL import Image
im = Image.open("/Users/stephaniexia/Documents/didi/yinyuehui.jpg")

#If successful, this function returns an Image object. 
#You can now use instance attributes to examine the file contents.

print("format:",im.format,"  size:", im.size,"  mode", im.mode)

# mode:Common modes are “L” (luminance) for greyscale images, 
#“RGB” for true colour images, and “CMYK” for pre-press images.
```

```python
'''
Once you have an instance of the Image class, you can use the methods defined by this class
to process and manipulate the image. For example, let’s display the image we just loaded:
The standard version of show is not very efficient,
since it saves the image to a temporary file and calls the xv utility to display the image.
If you don’t have xv installed, it won’t even work. When it does work though, 
it is very handy for debugging and tests. xv是一个应用于X-Window系统的互动式图像处理程序
'''
im.show()
```

```python
from pylab import *
pil_im = Image.open("/Users/stephaniexia/Documents/didi/yinyuehui.jpg")
figure()
subplot(121)
title("origin:")
axis('off')
imshow(pil_im)

#“L” (luminance) for greyscale images
pil_im_g = Image.open('/Users/stephaniexia/Documents/didi/yinyuehui.jpg').convert('L')
gray()
subplot(122)
title("grey")
axis('off')
imshow(pil_im_g)
```

```python
# change format using save():
import os

infile = "/Users/stephaniexia/Documents/didi/yinyuehui.jpg"
outfile = os.path.splitext(infile)[0]+".png"
print(outfile)
try:
    Image.open(infile).save(outfile)
except IOError:
    print("cannot convert", infile)
```

```python
#copy and paste region:

pil_im = Image.open('/Users/stephaniexia/Documents/didi/yinyuehui.jpg')
box = (100,100,400,400)
region = pil_im.crop(box)
region = region.transpose(Image.ROTATE_180)
pil_im.paste(region,box)
title('copy&paste')
axis('off')
imshow(pil_im)
```

```python
# thumbnail
pil_im = Image.open('/Users/stephaniexia/Documents/didi/yinyuehui.jpg')
size = 128, 128
pil_im.thumbnail(size)
print(pil_im.size)
title('thumbnail:')
axis('off')
imshow(pil_im)
pil_im.save('/Users/stephaniexia/Documents/didi/yinyuehuithumbnail.jpg')
```

```python
# resize:
pil_im = Image.open('/Users/stephaniexia/Documents/didi/yinyuehui.jpg')
pil_im = pil_im.resize(size)
print(pil_im.size)
title('resize:')
axis('off')
imshow(pil_im)
```

```python
# rotate 45°
pil_im = Image.open('/Users/stephaniexia/Documents/didi/yinyuehui.jpg')
pil_im = pil_im.rotate(45)
title('rotate 45°:')
axis('off')
imshow(pil_im)

show()
```
